```python
# 10. Tabulate and visualize the years and countrywide tweets of the sentiments. Pick tweets with sentiments, â€˜Positive,â€™ â€˜Negative,â€™ and â€˜Neutralâ€™ only.
```

```python
filtered_df = df[df['Sentiment'].isin(['positive', 'negative', 'neutral'])]

sentiment_counts = filtered_df.groupby(['Year', 'Country'])['Sentiment'].value_counts().unstack(fill_value=0)

sentiment_counts.plot(kind='bar', figsize=(15, 8))
plt.title('Countrywide Tweet Sentiments Over Years')
plt.xlabel('Year, Country')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()


sentiment_counts
```
