```python
stars = df['Star Cast'].str.split('(?<=[a-z])(?=[A-Z])', expand=True).stack()
stars = stars.reset_index(drop=True)
star_counts = stars.value_counts().reset_index()
star_counts.columns = ['Star', 'Count']

genre_counts = df['Genre'].str.split(',').explode().str.strip()
genre_star_counts = genre_counts.value_counts().reset_index()
genre_star_counts.columns = ['Genre', 'Counts']

plt.figure(figsize=(12, 6))
plt.bar(genre_star_counts['Genre'], genre_star_counts['Counts'], color='skyblue')
plt.title('Number of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

year_counts = df['Year'].value_counts().reset_index()
year_counts.columns = ['Year', 'Counts']

plt.figure(figsize=(12, 6))
plt.bar(year_counts['Year'], year_counts['Counts'], color='lightgreen')
plt.title('Number of Movies by Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

certificate_counts = df['Certificates'].value_counts().reset_index()
certificate_counts.columns = ['Certificate', 'Counts']

plt.figure(figsize=(12, 6))
plt.bar(certificate_counts['Certificate'], certificate_counts['Counts'], color='salmon')
plt.title('Number of Movies by Certificate')
plt.xlabel('Certificate')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Top Stars by Number of Movies:")
print(star_counts.head(10))
```

```python
# Comparative analysis of IMDB ratings and the metascore
```

```python
data = df.groupby('IMDb Rating', as_index=False)['MetaScore'].mean()
plt.figure(figsize=(8, 6))
sns.lineplot(data=data, x='IMDb Rating', y='MetaScore', marker='o')
plt.title('Line Graph of Average MetaScore by IMDb Rating')
plt.xlabel('IMDb Rating')
plt.ylabel('Average MetaScore')
plt.grid()
plt.show()
```

```python
# Analyse the relationship between movie attributes and ratings.
```

```python
heatmap_data = df[['IMDb Rating', 'MetaScore', 'Duration (minutes)']]
correlation_matrix = heatmap_data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of IMDb Dataset')
plt.show()
```
