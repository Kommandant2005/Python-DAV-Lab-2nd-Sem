```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Week 6 , Exercise 2: Dataset : MTCARS
```

```python
data=pd.read_csv("mtcars.csv",index_col=0)
```

```python
data
```

```python
# 1. Explore the distribution of Fuel efficiency of the cars
data["mpg"].plot(kind='hist', bins=10, title="Histogram of th Distribution of MPG")
```

```python
# 2. Find the cars with the best fuel efficiency
mpg=data["mpg"].max()
data[data["mpg"]==mpg]
```

```python
mpg=data["mpg"].min()
data[data["mpg"]==mpg]
```

```python
# 3. How many cars are manual vs automatic?
data["am"].value_counts()
```

```python
# 4. Find the car with the worst horsepower  
min_hp=data['hp'].min()
data[data['hp']==min_hp]
```

```python
# 5. Find 5 number summary and draw the boxplot of displacement. 
data.describe()
```

```python
data["disp"].plot(kind='box',title='Histogram of th Distribution of Displacement')
```

```python
# 6. Which is the heaviest car? How many gears does it have?
max_wt=data['wt'].max()
data[data['wt']==max_wt]
```

```python
# 7. Which is the car with the best qsec?
max_qsec=data['qsec'].max()
data[data['qsec']==max_qsec]
```

```python
# 8. What is average MPG for manual vs. automatic cars? 
dat_man= data[data["am"]==1]["mpg"]
print("Average MPG of manual cars:",dat_man.mean())
dat_auto= data[data["am"]==0]["mpg"]
print("Average MPG of automatic cars:",dat_auto.mean())
```

```python
# 9. Draw Side by Side box plot to understand the difference in fuel efficience of Manual vs Automatic cars.  Analyze and write about fuel efficiency in each group (manual vs. automatic). 
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
data[data['am'] == 1]['mpg'].plot(kind='box')
plt.title('Manual Cars')
plt.ylabel('Fuel Efficiency (mpg)')

plt.subplot(1, 2, 2)
data[data['am'] == 0]['mpg'].plot(kind='box')
plt.title('Automatic Cars')
plt.ylabel('Fuel Efficiency (mpg)')
plt.show()
```

```python
# 10. What is the relationship between the weight of the car and MPG?
dat_x= data[["mpg","wt"]]
dat_x.plot(kind='scatter', x='mpg', y='wt', color='blue', title='Scatter Plot from DataFrame')
```

```python
# 11. Categorize the cars based on the number of gears in the cars. How many cars are there in each type?
data['gear'].nunique()
data['gear'].value_counts()
```

```python
# 12. What is the relationship between fuel efficiency and the number of gears in the car? 
data[data['gear'] == 3]['mpg'].plot(kind='bar',color='purple')
plt.title('3 geared Car')
plt.ylabel('Fuel Efficiency (mpg)')
```

```python
data[data['gear'] == 4]['mpg'].plot(kind='bar',color='purple')
plt.title('4 geared Car')
plt.ylabel('Fuel Efficiency (mpg)')
```

```python
data[data['gear'] == 5]['mpg'].plot(kind='bar',color='purple')
plt.title('5 geared Cars')
plt.ylabel('Fuel Efficiency (mpg)')
```

```python
# 13. Explain the relationship between horse power and number of cylinders in the car.
data[data['cyl'] == 4]['hp'].plot(kind='bar',color='pink')
plt.title('4 Cylinder Cars')
plt.ylabel('Horse Power')
```

```python
data[data['cyl'] == 6]['hp'].plot(kind='bar',color='pink')
plt.title('6 Cylinder Cars')
plt.ylabel('Horse Power')
```

```python
data[data['cyl'] == 8]['hp'].plot(kind='bar',color='pink')
plt.title('8 Cylinder Cars')
plt.ylabel('Horse Power')
```

```python
# 14. Explain the relationship between displacement and gross horse power. 
dat_x= data[["disp","hp"]]
dat_x.plot(kind='scatter', x='disp', y='hp', color='blue', title='Scatter Plot from Displacement vs Horse Power')
```

```python
# 15. Which car would I pick if I am looking for high speed as well as good fuel efficiency?
dat_x= data[["mpg","qsec"]]
dat_x.plot(kind='scatter', x='mpg', y='qsec', color='red', title='Scatter Plot from Fuel Efficiency vs Speed')
```

```python
# Week 6 , Exercise 2: Dataset : CEREALS
```

```python
cereals=pd.read_csv("cereals.csv",index_col=0)
```

```python
numeric_cols = cereals.select_dtypes(include=['number']).columns
for col in numeric_cols:
    non_zero_mean = cereals.loc[cereals[col] != 0, col].mean()
    cereals.loc[cereals[col] == 0, col] = non_zero_mean
```

```python
cereals
```

```python
# 1. How many cereals(unique brands) are there in the dataset?
print("Number of Unique Brands of Cereals Available")
cereals['mfr'].nunique()
```

```python
# 2. How many cereals are listed per manufacturer of the cereal? 
cereals['mfr'].value_counts()
```

```python
# 3. How many hot vs cold cereals? 
cereals['type'].value_counts()
```

```python
# 4. Which is the best and worst cereal? 
print("Worst Rated Cereal")
min_rating=cereals["rating"].min()
cereals[cereals["rating"]==min_rating]
```

```python
print("Best Rated Cereal")
max_rating=cereals["rating"].max()
cereals[cereals["rating"]==max_rating]
```

```python
# 5. Compare the rating for hot vs cold cereals. Which do people prefer? 
cereals[cereals['type'] == 'C']['rating'].plot(kind='box',color='yellow')
plt.title('Cold Cereals')
plt.ylabel('Rating')
```

```python
cereals[cereals['type'] == 'H']['rating'].plot(kind='box',color='yellow')
plt.title('Hot Cereals')
plt.ylabel('Rating')
```

```python
# 6. Which cereals have the highest fiber and lowest sugar? 
dat_x= cereals[["fiber","sugars"]]
dat_x=dat_x.reset_index(drop=True)
dat_x.plot(kind='scatter', x='fiber', y='sugars', color='red', title='Scatter Plot from Fiber Content vs Sugar')
```
