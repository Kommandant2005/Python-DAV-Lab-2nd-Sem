```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('deliveries.csv')
df
```

```python
# 1. Find and visualize the total runs scored by each team.
```

```python
total_runs_scored = df.groupby('batting_team')['total_runs'].sum()
plt.figure(figsize=(10, 6))
plt.bar(total_runs_scored.index, total_runs_scored.values)
plt.xlabel('Team')
plt.ylabel('Total Runs Scored')
plt.title('Total Runs Scored by Each Team')
plt.xticks(rotation=45)
plt.show()
```

```python
#2. Identify and display the names of the top 10 highest-scoring batters with their total runs scored.
```

```python
top_10_batters = df.groupby('batter')['total_runs'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10_batters.index, top_10_batters.values, color = 'lightgreen')
plt.xlabel('Runs Scored')
plt.ylabel('Batsman')
plt.title('Runs Scored by each batsman')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Find and visualize the top 10 bowlers with the most wickets.
```

```python
top_10_bowlers = df.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_10_bowlers.index, top_10_bowlers.values, color = 'pink')
plt.xlabel('Wickets Taken')
plt.ylabel('Bowler')
plt.title('Wickets Taken by each Bowler')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Tabulate and visualize the total number of extras conceded by each team.
```

```python
total_extras = df.groupby('batting_team')['extra_runs'].sum()
plt.figure(figsize=(10, 6))
plt.bar(total_extras.index, total_extras.values, color = 'gold')
plt.xlabel('Batting team')
plt.ylabel('Extra Runs Conceded')
plt.title('Extra Runs Scored by Each Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Tabulate and visualize the batsmen (top 5) involved in the most run-outs.
```

```python
df['dismissal_kind'].unique()
```

```python
run_out_dismissals = df[df['dismissal_kind'] == 'run out']
run_out_dismissals=run_out_dismissals['player_dismissed'].value_counts().head(5)

plt.figure(figsize=(10, 6))
plt.plot(run_out_dismissals.index, run_out_dismissals.values)
plt.xlabel('Batsman')
plt.ylabel('Runouts')
plt.title('Runouts for each batsman')
plt.xticks(rotation=45)
plt.show()

run_out_dismissals
```

```python
# 6. Tabulate and visualize the top 5 bowlers who conceded the most extras.
```

```python
bowlers_5 = df.groupby('bowler')['extra_runs'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.plot(bowlers_5.index, bowlers_5.values)
plt.xlabel('Bowlers')
plt.ylabel('Extra runs')
plt.title('Extra runs for each bowler')
plt.xticks(rotation=45)
plt.show()

bowlers_5
```

```python
# 7. Tabulate and visualize the most common dismissal type.
```

```python
dismissal_counts = df['dismissal_kind'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(dismissal_counts.values, labels=dismissal_counts.index, autopct='%1.1f%%', startangle=90)
plt.xlabel('Dismissal Type')
plt.ylabel('Count')
plt.title('Most Common Dismissal Types')
plt.show()
dismissal_counts
```

```python
# 8. Find and display which of the top 3 overs has the highest run rate.
```

```python
top_3_overs = df[df['over'] <= 3]
run_rate_by_over = top_3_overs.groupby('over')['total_runs'].mean()

plt.figure(figsize=(10, 6))
plt.bar(run_rate_by_over.index, run_rate_by_over.values)
```

```python
# 9. Analyze and visualize powerplay performance (first 6 overs).
```

```python
powerplay_df = df[df['over'] <= 6]
powerplay_runs = powerplay_df.groupby('over')['total_runs'].sum()

plt.figure(figsize=(10, 6))
plt.plot(powerplay_runs.index, powerplay_runs.values, marker='o', linestyle='-', color='b')
```

```python
# 10. Tabulate and visualize the top 5 batsmen with the most sixes and fours.
```

```python
sixes = df[df['batsman_runs'] == 6].groupby('batter')['batsman_runs'].count()
fours = df[df['batsman_runs'] == 4].groupby('batter')['batsman_runs'].count()

batsman_boundaries = pd.DataFrame({'Sixes': sixes, 'Fours': fours}).fillna(0)
batsman_boundaries['Total Boundaries'] = batsman_boundaries['Sixes'] + batsman_boundaries['Fours']

top_5_batsman = batsman_boundaries.sort_values(by='Total Boundaries', ascending=False).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_5_batsman.index, top_5_batsman['Sixes'], label='Sixes')
plt.bar(top_5_batsman.index, top_5_batsman['Fours'], bottom=top_5_batsman['Sixes'], label='Fours')
plt.xlabel('Batsman')
plt.ylabel('Number of boundaries')
plt.title('Top 5 Batsmen with the most Sixes and Fours')
plt.xticks(rotation=45)
plt.legend()
plt.show()

top_5_batsman
```
